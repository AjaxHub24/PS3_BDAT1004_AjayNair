```python
import pandas as pd
import numpy as np
```

```python
data_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
user_data = pd.read_csv(data_url, sep='|')
```

```python
mean_age_by_occupation = user_data.groupby('occupation')['age'].mean()
print("Mean Age per Occupation:")
print(mean_age_by_occupation)    # AVG AGE BY OCCUPATION
```

```python
male_ratio_by_occupation = user_data[user_data['gender'] == 'M'].groupby('occupation')['gender'].count() / user_data.groupby('occupation')['gender'].count()
sorted_male_ratio = male_ratio_by_occupation.sort_values(ascending=False)
print("\nMale Ratio per Occupation (Sorted):\n")
print(sorted_male_ratio)
```

```python
age_range_by_occupation = user_data.groupby('occupation')['age'].agg(['min', 'max'])
print("\nAge Range per Occupation:\n")
print(age_range_by_occupation)   #AGE RANGE
```

```python
mean_age_by_occupation_and_sex = user_data.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age per Occupation and Sex:\n")
print(mean_age_by_occupation_and_sex)    #MEAN AGE/OCCUPATION AND SEX
```

```python
gender_distribution_by_occupation = user_data.groupby(['occupation', 'gender'])['gender'].count() / user_data.groupby('occupation')['gender'].count() * 100
print("\nGender Distribution per Occupation:\n")
print(gender_distribution_by_occupation)     #PERCENTAGE OF WOMEN AND MEN IN OCCUPATION
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('wind.txt', delim_whitespace=True)
```

```python
data['Date'] = pd.to_datetime(data['Dy'].astype(str) + '/' + data['Mo'].astype(str) + '/' + data['Yr'].astype(str), format='%d/%m/%y')
data.set_index('Date', inplace=True)
data.drop(['Dy', 'Mo', 'Yr'], axis=1, inplace=True)
```

```python
def fix_year(year):
    if year > 2000:
        return year - 100
    return year

data.index = data.index.to_series().apply(lambda x: x.replace(year=fix_year(x.year)))
```

```python
data.index = pd.to_datetime(data.index)
data.index.name = 'Date'
```

```python
missing_per_location = data.isnull().sum()
print("Missing values per location:\n", missing_per_location)
```

```python
non_missing_total = data.notnull().sum().sum()
print("Total non-missing values:", non_missing_total)
```

```python
mean_windspeed = data.mean().mean()
print("Mean windspeeds over all locations and times:", mean_windspeed)
```

```python
loc_stats = data.describe()
print("Location statistics:\n", loc_stats)
```

```python
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])
print("Location statistics:\n", loc_stats)
```

```python
january_data = data.resample('M').mean()  # Resample to monthly frequency
january_data = january_data[january_data.index.month == 1]  # Filter January data
january_avg = january_data.groupby(january_data.index.year).mean()
print("Average windspeed in January:\n", january_avg)
```

```python
yearly_data = data.resample('Y').mean()
print("Yearly downsampled data:\n", yearly_data)
```

```python
monthly_data = data.resample('M').mean()
print("Monthly downsampled data:\n", monthly_data)
```

```python
weekly_data = data.resample('W').mean()
print("Weekly downsampled data:\n", weekly_data)
```

```python
weekly_stats = weekly_data.iloc[1:53].describe()
print("Weekly statistics for the first 52 weeks:\n", weekly_stats)
```

QUESTION 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = [1900, 1950, 2000]
subset = data[data['Year'].isin(years)].copy()  
subset.loc[:, 'Marriages_per_capita'] = subset['Marriages'] / subset['Population']
subset.loc[:, 'Divorces_per_capita'] = subset['Divorces'] / subset['Population']
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(subset['Year'], subset['Marriages_per_capita'], label='Marriages', color='skyblue')
ax.bar(subset['Year'], subset['Divorces_per_capita'], label='Divorces', color='lightcoral')
ax.set_xlabel('Year')
ax.set_ylabel('Number per capita')
ax.set_title('Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
ax.legend()
plt.show()
```

QUESTION 9 

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('roman-emperor-reigns.csv')

# Group by cause of death and count the number of emperors
death_cause_counts = df['Cause_of_Death'].value_counts()

# Define custom colors for the pie chart
pie_colors = ['#ff69b4', '#33aaff', '#33ff99', '#ffa07a', '#b2b2f7', '#ffa6c9', '#b2f7b2', '#ffa0a0']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    death_cause_counts,
    labels=death_cause_counts.index,
    autopct='%.1f%%',
    startangle=140,
    explode=[0.01] * len(death_cause_counts),
    colors=pie_colors,
    textprops={'fontsize': 17}
)

# Set title with custom formatting
plt.title("Cause of Death", fontsize=22, fontweight='bold', color='blue')

# Display the pie chart
plt.show()
```

Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

plt.figure(figsize=(8, 8))

scatter = plt.scatter(data['Total Arcade Revenue (billions)'], data['Computer Science Doctorates Awarded (US)'],
            c=data['Year'], cmap='viridis', s=100, edgecolors='black', alpha=0.8) 

plt.title('Relationship between Arcade Revenue and CS Doctorates (2000-2009)')
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.colorbar(scatter, label='Year')
plt.tight_layout()
plt.show()
```
